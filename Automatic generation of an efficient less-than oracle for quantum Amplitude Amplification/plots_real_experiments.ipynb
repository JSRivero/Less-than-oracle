# Generation of the plots in paper

```python
import pandas as pd

import matplotlib.pyplot as plt
import os
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=18)
```

# Unitary Gate

```python
number = 4
nqubits = 4
```

```python
df_unitary = pd.read_csv('data/less_than_4_with_4_qubits_2e4_shots_unitary_REAL_EXPERIMENT.csv', index_col=0).squeeze()
```

```python
result = dict(zip([int(str(x),2) for x in df_unitary.index.tolist()], df_unitary.values.tolist()))
```

```python
x = list(range(2**nqubits))
y = [result.get(a, 0) for a in x]
```

```python
xlabel = 'Numbers ($k$)'
ylabel = 'Frequency'

edgecolor = 'k'
figsize_ = (25,6)

# Save figure
folder = 'images'
name_image = 'less_than_%d_with_%d_qubits_2e4_shots_unitary_REAL_EXPERIMENT_CORRECTION.png'%(number, nqubits)
dpi_ = 400

#Legend
legend_names = ['desired states','undesired states']
legend_names = ['$k<%d$'%number, r'$k\geq %d$'%number]

fontsize = 40
fontsize_ticks = 40

# Label and ticks

yticks_size = fontsize_ticks

xticks_size = fontsize_ticks
xlabel_size = fontsize
ylabel_size = fontsize

# xlim
xlim = [-1, 16]

# ylabel_position 
ylabel_position = (-0.065, 0.5)

# Title
grid = True
```

```python
x_desired = x[:number]
x_undesired = x[number:]
y_desired = y[:number]
y_undesired = y[number:]
```

```python
fig, ax = plt.subplots(1, 1, figsize=figsize_)

n_shots_plot = sum(y)

ax.bar(x_desired, [y_element/n_shots_plot for y_element in y_desired], zorder=3,
       edgecolor='k', label=legend_names[0], color = 'royalblue', linewidth=6)
ax.bar(x_undesired, [y_element/n_shots_plot for y_element in y_undesired], zorder=3,
       edgecolor='k', label=legend_names[1], color='tomato', linewidth=0.8)


ax.tick_params(axis='x', labelsize=xticks_size)
ax.tick_params(axis='y', labelsize=yticks_size)

ax.set_xlabel(xlabel, fontsize=xlabel_size)
ax.set_ylabel(ylabel, fontsize=ylabel_size)

ax.set_xlim(xlim)

ax.yaxis.set_label_coords(ylabel_position[0], ylabel_position[1])

ax.grid(axis='y', zorder=0)

ax.set_xticks(sorted(x))
# ax.set_xticklabels([str(a) if a%2==0 else ' ' for a in range(2**nqubits)])
    
ax.legend(fontsize=fontsize)

ax.tick_params(width=2, size=8)

plt.savefig(os.path.join(folder, name_image), dpi=dpi_, bbox_inches='tight')
```
